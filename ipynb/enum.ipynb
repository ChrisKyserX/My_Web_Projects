#### 使用 #!share 指令 显式地把前面的变量“导入”到当前单元格。
#### 使用 #!set --preserve 让整个会话持久化

在每个代码单元格的**第一行**（必须是第一个非空字符）写上 magic command：

| 语言 | Magic Command | 示例 |
| :--- | :--- | :--- |
| **C#** | `#!csharp` | `#!csharp`<br>`Console.WriteLine("Hello");` |
| **F#** | `#!fsharp` | `#!fsharp`<br>`printfn "Hello"` |
| **PowerShell** | `#!pwsh` 或 `#!powershell` | `#!pwsh`<br>`Get-Process` |
| **JavaScript** | `#!javascript` 或 `#!js` | `#!javascript`<br>`console.log("Hello");` |
| **SQL** | `#!sql` | `#!sql`<br>`SELECT * FROM users` |
| **HTML** | `#!html` | `#!html`<br>`<h1>Title</h1>` |
| **Python** | `#!python` | `#!python`<br>`print("Hello")` |
| **R** | `#!r` | `#!r`<br>`print("Hello")` |
